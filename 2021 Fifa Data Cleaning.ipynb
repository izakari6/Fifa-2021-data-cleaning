```python
#lets clean the club column
c_club = df["Club"].str.replace("\n","")
df["Club"] = c_club
```

```python
#clean positions
n_pos = df["Positions"].str.replace(",", "")

df["Positions"] = n_pos
```

```python
#lets clean height
# height has 2 units of measurement thus cm and ft_inch
# transfor all into centimeters

df["Height"] = df["Height"].astype("string")

cm_height = []

for i, height in enumerate(df["Height"]):
#removing cm unit
    if "cm" in height:
        cm_value = float(height.replace("cm", ""))
        
    else:
        ft_inch = height.replace("\\","").replace("'",".").replace('"',"").split(".")
        ft,inch = float(ft_inch[0]), float(ft_inch[1])
        cm_value = (ft * 30.48) + (inch * 2.54)
    
    cm_height.append(round(cm_value,2))
    
df["Height"] = cm_height
```

```python
#Weight
#also has same issue as height
# let standardize all kg as lbs

df["Weight"] = df["Weight"].astype("string")

lbs_weight = []

#converting kg to lbs
for i, weight in enumerate(df["Weight"]):
    if "kg" in weight:
        kg_value = float(weight.replace("kg", ""))
        lbs_value = kg_value * 2.204
    else:
        lbs_value = float(weight.replace("lbs",""))
    
    lbs_weight.append(round(lbs_value,2))

df["Weight"] = lbs_weight
```

```python
#Value
#remove currency sign and value abbrev.

n_val = []
for i in range(len(df["Value"])):
        v = df["Value"][i]
        v = v.replace("€", "")
        v = v.replace("K","000")
        v = v.replace("M","000_000")
        v = v.replace(".","")     
        n_val.append(int(v))
        
df["Value"] = n_val      
```

```python
#Wage
#remove currency sign and value abbrev.

n_wage = []

for i in range(len(df["Wage"])):
    w = df["Wage"][i]
    w = w.replace("€", "")
    w = w.replace("K","000")
    w = w.replace("M","000_000")
    w = w.replace(".","")     
    n_wage.append(int(w))
    
df["Wage"] = n_wage
```

```python
# release clause
#remove currency sign and value abbrev.

n_claus = []

for i in range(len(df["Release Clause"])):
    c = df["Release Clause"][i]
    c = c.replace("€", "")
    c = c.replace("K","000")
    c = c.replace("M","000_000")
    c = c.replace(".","") 
    n_claus.append(int(c))
    
df["Release Clause"] = n_claus
```

```python
#fill nulls with 0

n_ldate = df["Loan Date End"].fillna(0)

#create a copy 
n_ldate_copy = n_ldate

#iterate to create our new data column
for i in n_ldate_copy:
    contract = pd.Series(n_ldate_copy.apply(lambda x: "Not on loan" if i == 0 else "On loan" ))

df["Contract Type"] = contract
```

```python
# W/F
#since we have just few unique data we can set it as categorical data and replace the special characters

wf = []

for i in range(len(df["W/F"])):
    w = df["W/F"][i]
    w = int(w.replace("★","").strip())
    wf.append(w)

df["W/F"] = wf
df["W/F"] = pd.Categorical(df["W/F"],ordered = True)
```

```python
# IR
#since we have just few unique data we can set it as categorical data and replace the special characters

df["IR"] = df["IR"].astype("string")
ir = []

for i in range(len(df["IR"])):
    i = df["IR"][i]
    i = int(i.replace("★","").strip())
    ir.append(i)
    
df["IR"] = ir
df["IR"] = pd.Categorical(df["IR"], ordered = True)
```

```python
#SM
#since we have just few unique data we can set it as categorical data and replace the special characters

df["SM"] = df["SM"].astype("string")

sm = []

for i in range(len(df["SM"])):
    s = df["SM"][i]
    s = int(s.replace("★","").strip())
    sm.append(s)

df["SM"] = sm
df["SM"] = pd.Categorical(df["SM"], ordered = True)
```

```python
#Hits
# HIts has some null cells which is almost 14% of the cells
#depending on the purpose of the analysis, we could fill the null with either 0 or peform interpolation
#also some values have char within it
#lets fix it all
#df["Hits"] = df["Hits"].astype("int")

df["Hits"].isnull().sum()
```

```python
hits =[]

df["Hits"] = df["Hits"].fillna(0)
df["Hits"] = df["Hits"].astype(str)


for i in range(len(df["Hits"])):
    h = df["Hits"][i]
    h = h.replace(".","")
    h = h.replace("k","000")
    h = h.replace("K","000")
    hits.append(int(h))

df["Hits"] = hits
```

```python
#Contract column has really messy data
#best way is to sort it into different columns

def contract_sort(Contract):
    if Contract == "Free":
        c_type = "Free Agent"
        c_start = None
        c_end = None
    elif " On Loan" in Contract:
        c_type = "On Loan"
        c_start = None
        c_end = Contract.replace(" On Loan", "")
    else:
        c_type = "On Contract"
        c_start, c_end = Contract.split(" ~ ")
        
    return c_start, c_end, c_type

df[["Contract Start", "Contract End", "ContractType"]] = df["Contract"].apply(lambda x : pd.Series(contract_sort(x)))   
```

```python
df["ContractType"].unique()

#since Contract type has fewer unique variable let make it a categoral data
df["ContractType"] = pd.Categorical(df["ContractType"])
```

```python
#A/W has fewer unique data so we can set it as a categorical data

df["A/W"].unique()
df["A/W"] = pd.Categorical(df["A/W"], ['Low','Medium','High'], ordered = True)
```

```python
#D/W has fewer unique data so we can set it as a categorical data

df["D/W"].unique()
df["D/W"] = pd.Categorical(df["D/W"], categories = ['Low','Medium','High'], ordered = True)
```
